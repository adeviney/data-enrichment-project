```python
import json
with open('/Users/Alexis/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# import the YelpAPI Class
from yelpapi import YelpAPI
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='San Francisco, CA 94104',
                                       term='Vegan')
print(type(search_results))
search_results.keys()
```

```python
import pandas as pd
biz1 = pd.DataFrame(search_results['businesses'])
biz1.head()
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='San Francisco, CA 94104',
                                       term='Vegan',
                                      offset = 20)
```

```python
biz20 = pd.DataFrame(search_results['businesses'])
biz20.head()
```

```python
## concatenate the previous results and new results. 
businesses = pd.concat([biz1, biz20],
                      ignore_index=True)
display(businesses.head(3), businesses.tail(3))
```

```python
businesses.to_json('veganSFfood.json', orient='records')
```
